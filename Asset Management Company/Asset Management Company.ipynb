11.Identify the top 9 countries in terms of highest investment in ‘venture’ fund type.


```python

venture_data_table = master_table[master_table['funding_round_type'] == 'venture']


country = venture_data_table.groupby('country_code')['raised_amount_usd'].sum()


sorted_country = country.sort_values(ascending=False)


top_9_countries = sorted_country.head(9)


print(top_9_countries)
```

13. Using the user defined function convert the ‘rasied_amount_used’ column into a categorical
column as followa) If the amount is less than 5 Millions then recode as TypeA.
b) If the amount is greater than 5 Millions and less than 7 Millions then recode as TypeB.
c) If the amount is greater than 7 Millions then recode as TypeC

```python
import pandas as pd


def convert_raised_amount(amount):
    if amount < 5000000:
        return 'TypeA'
    elif 5000000 <= amount < 7000000:
        return 'TypeB'
    else:
        return 'TypeC'





master_table['raised_amount_category'] = master_table['raised_amount_usd'].apply(convert_raised_amount)


master_table
```

```python
## converted the ‘rasied_amount_used’ column into a categorical column with 3 categories.
```
